```python
import pandas as pd
#df1 is for wheat
df1 = pd.read_excel("/content/Ukraine Russia Food Insights - ukraine-russia-food.xlsx",sheet_name = "wheat")
#df2 is for maize
df2 = pd.read_excel("/content/Ukraine Russia Food Insights - ukraine-russia-food.xlsx",sheet_name = "maize")
#df3 is for barley
df3 = pd.read_excel("/content/Ukraine Russia Food Insights - ukraine-russia-food.xlsx",sheet_name = "barley")
#df4 is for sunflower
df4 = pd.read_excel("/content/Ukraine Russia Food Insights - ukraine-russia-food.xlsx",sheet_name = "sunflower")
```

```python
#checking for missing values
missing_values = df1.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
#checking datatype
data_types = df1.dtypes
print("\nData Types:\n", data_types)
```

```python
#checking for duplicate values
duplicates = df1.duplicated()
print("\nDuplicate Rows:\n", df1[duplicates])
```

```python
#handling columns which can be replaced with zero

df1 = df1.fillna({
    "Wheat exports": 0,
    "Wheat imports": 0,
    "Wheat imports (% domestic supply)": 0,
    "Wheat exports per capita":0,
    "Wheat imports per capita":0,
    "Wheat imports from Ukraine":0,
    "Wheat imports from Russia":0,
    "Wheat imports from Ukraine per capita":0,
    "Wheat imports from Russia per capita":0,
    "Wheat imports from Ukraine (% imports)":0,
    "Wheat imports from Russia (% imports)":0,
    "Wheat imports from Ukraine (% supply)":0,
    "Wheat imports from Russia (% supply)":0
})
```

```python
#handling other columns which can't be zero

df1["Net wheat imports"] = df1["Wheat imports"]-df1["Wheat exports"]
df1["Net wheat imports per capita"] = df1["Wheat imports per capita"] - df1["Wheat exports per capita"]
df1["Wheat imports from Ukraine + Russia"] = df1["Wheat imports from Ukraine"]+df1["Wheat imports from Russia"]
df1["Wheat imports from Ukraine + Russia per capita"]=df1["Wheat imports from Ukraine per capita"]+df1["Wheat imports from Russia per capita"]
df1["Wheat imports from Ukraine + Russia (% imports)"]=df1["Wheat imports from Ukraine (% imports)"]+df1["Wheat imports from Russia (% imports)"]
df1["Wheat imports from Ukraine + Russia (% supply)"]=df1["Wheat imports from Ukraine (% supply)"]+df1["Wheat imports from Russia (% supply)"]
```

```python
#checking for missing values
missing_values = df1.isnull().sum()
print("Missing Values:\n", missing_values)
```
